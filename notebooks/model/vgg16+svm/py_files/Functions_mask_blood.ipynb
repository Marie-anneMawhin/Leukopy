```python
def process_image(img1):
    
    central_contour_found=0
    
    img_grey2=(img1[:,:,0]+img1[:,:,2])/(2*img1[:,:,1])  #RGB

    img_grey5=img_grey2.copy()
    for i in range(img_grey5.shape[0]):
        for j in range(img_grey5.shape[1]):

                if img_grey2[i][j]<5:    #5
                          img_grey5[i,j]=0

    ret, thresh = cv2.threshold(img_grey5.astype(np.uint8),0,255,cv2.THRESH_BINARY_INV)  


    kernel = np.ones((3,3),np.uint8)
    opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 25)

    # sure background area
    sure_bg = cv2.dilate(opening,kernel,iterations=10)

    des = cv2.bitwise_not(sure_bg)
    contour,hier = cv2.findContours(des,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)

    for cnt in contour:
        cv2.drawContours(des,[cnt],0,255,-1)

    sure_bg = cv2.bitwise_not(des)


    img_grey7=np.ones(sure_bg.shape)*255
    ret, thresh2 = cv2.threshold(sure_bg, 127, 255, 0)
    contours, hierarchy = cv2.findContours(thresh2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    image=cv2.drawContours(img_grey7, contours, -1, (0, 255, 0), 1)

    img_grey8=np.ones(sure_bg.shape)*255
    #chercher le contour central
    m=[]
    for ip in range(len(contours)):
        x=[contours[ip][:,:,0].mean(),contours[ip][:,:,1].mean()]
        m.append(np.sqrt((x[0]-sure_bg.shape[0]/2)**2+ (x[0]-sure_bg.shape[0]/2)**2))


    
    
    if len(m)>1:
        m=m[1:]
        contour_center=contours[np.argmin(m)+1]

        image2=cv2.drawContours(img_grey8, [contour_center], -1, (0, 255, 0), 1)
        dist_transform = cv2.distanceTransform(image2.astype(np.uint8),cv2.DIST_L2,5)

        ret, sure_fg = cv2.threshold(dist_transform,0.1*dist_transform.max(),255,0)
        sure_fg = ndimage.binary_fill_holes(255-sure_fg).astype(int)
        img6=img1.copy()                 
        for k in [0,1,2]:
           for i in range(img6.shape[0]):

              for j in range(img6.shape[1]):
                 if sure_fg [i,j]==0 :  #sure_bg
                      img6[i,j,k] =0 
                        
        return 1,img6   # on a trouvé bonne image
        
    elif len(m)==1:
        contour_center=contours[np.argmin(m)]

        image2=cv2.drawContours(img_grey8, [contour_center], -1, (0, 255, 0), 1)
        dist_transform = cv2.distanceTransform(image2.astype(np.uint8),cv2.DIST_L2,5)

        ret, sure_fg = cv2.threshold(dist_transform,0.01*dist_transform.max(),255,0)
        sure_fg = ndimage.binary_fill_holes(255-sure_fg).astype(int)
        img6=img1.copy()                 
        for k in [0,1,2]:
           for i in range(img6.shape[0]):

             for j in range(img6.shape[1]):
                 if sure_fg [i,j]!=0 :  #sure_bg
                      img6[i,j,k] =0 
        
        
        
        
        return 1,img6   # on a trouvé bonne image
    else: 
        return 0,img1
            
```
